###Read CSV from API

```python
import pandas as pd

#https://raw.githubusercontent.com/mostafamos/demo_for_technopro/main/train.csv

### in this example im using GIT raw view instead of calling an API with requests
df = pd.read_csv("https://raw.githubusercontent.com/mostafamos/demo_for_technopro/master/train.csv")

df.head()

```

###Grouping and aggregating

```python
df = df.drop(["store","item"], axis=1)
df= df.groupby("date").sum().reset_index()
df.head()
```

###Locking data between two months

```python
df_two_cols = df.loc[:"2013-01-05"]
df_two_cols.head()
```

###Plot the data


```python
import matplotlib .pyplot as plt

plt.figure(figsize=(15,5))
plt.plot(df_two_cols['date'], df_two_cols['sales'])
plt.xlabel('Date')
plt.xlabel('Sales')
plt.title("Monthly Customer Sales")
plt.show()
```
